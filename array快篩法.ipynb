## 利用快篩畫圖

```python
X = np.linspace(-10,10,1000)
Y = np.sinc(X)
plt.plot(X,Y)
```

```python
plt.plot(X,Y)
plt.plot(X[Y<0],Y[Y<0],"o")
```
