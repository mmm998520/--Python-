```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.random.randn(4,3)
data
```

```python
df1 = pd.DataFrame(data,columns = list('abc'))
df1
```

```python
df2 = pd.DataFrame(np.random.randn(3,3),columns = list('abc'))
df2
```

## 用concat組合兩個DataFrame
需加入axis告知pandas組合的方向

index,columns可再次改變行列名稱

```python
df3 = pd.concat([df1,df2],axis = 0)
df3
```

```python
df3.index = range(7)
df3
```

```python
df4 = pd.concat([df1,df2],axis = 1)
df4
```

```python
df4.columns = list('abcdef')
df4
```
