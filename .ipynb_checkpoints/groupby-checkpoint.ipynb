```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("http://bit.ly/uforeports")
df.head()
```

## count計算個數

```python
df.count()
```

## groupby將同類群組
groupby+count計算群組內個數

```python
df_state = df.groupby("State").count()
df_state
```

## sort_values排序
ascending=False由大到小

```python
df_state.sort_values(by='Time',ascending=False)
```

```python
df_state
```

## inplace = True不建立直接更變DataFrame

```python
df_state.sort_values(by="Time",ascending = False,inplace = True)
df_state
```

```python
df_state.plot()
```

```python
df_state[:10].Time.plot(kind = 'bar')
```
