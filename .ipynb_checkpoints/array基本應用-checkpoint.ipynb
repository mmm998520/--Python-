## array強化randn功能
讓只能生成平均0,標準差1的randn做出變化

平均50，標準差8

```python
%matplotlib inline
import numpy as np
import matplotlib as plt
```

```python
a = np.array(np.random.randn(50))
a
```

```python
new_a = (a * 8) + 50
new_a
```

## 改變array的形狀(維度)
shape 直接改變造型
reshape 改變造型後賦予新的 array

```python
b = a.reshape(5,10)
print(a)
print(b)
```

```python
a.shape = (5,10)
a
```
