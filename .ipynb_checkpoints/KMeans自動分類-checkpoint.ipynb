## 當不知道數據如何分類時使用
讓機器分完類後再研究分類關係

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(100,2)
x[0]
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters = 3)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c = clf.labels_)
```
