```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
```

```python
x = np.linspace(0,2*np.pi,1000)
```

```python
def draw(n):
    y = np.sin(x*n)
    plt.plot(x,y,lw = 1)#lw可設定線寬
    plt.show()#因為jupyter內建快速畫圖，所以平常不需要，但現在我們要及時ren圖，所以需要
```

```python
draw(3)
```

```python
interact(draw,n=(0.,30.))
```
