```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## 畫出 f(x) = 1.2x + 0.8

```python
x = np.linspace(0,5,50)
x
```

```python
y = 1.2*x + 0.8
y
```

```python
plt.scatter(x,y)
plt.plot(x,y,'r')
```

## 給一個noise，讓他隨機一點，符合現實
f(x) + ε(x)

```python
x = np.linspace(0,5,50)
y = 1.2*x + 0.8 + 0.6 * np.random.randn(50)
i = 1.2*x + 0.8
```

```python
plt.scatter(x,y)
plt.plot(x,y,'b')
plt.plot(x,i,'r')
```

## 線性回歸機
將隨機點算出算性回歸

線性回歸機需要reshape，將資料做成單獨形式

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
x
```

```python
X = x.reshape(50,1)
X
```

```python
regr.fit(X,y)
```

```python
Y = regr.predict(X)
Y
```

```python
plt.scatter(x,y)
plt.plot(x,Y,'r')
plt.plot(x,i,'g')
```
