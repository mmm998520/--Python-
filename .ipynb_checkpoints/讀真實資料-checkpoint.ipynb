```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## 讀波士頓房價

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

## .feature_names查看內有的變數
可用 print(  .DESCR)查看變數功能

```python
boston.feature_names
```

```python
print(boston.DESCR)
```

```python
X = boston.data
Y = boston.target
```

```python
len(X)
```

```python
len(Y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,
                                                  test_size = 0.3,
                                                  random_state = 30)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
y_predict = regr.predict(x_test)
```

```python
plt.scatter(y_test,y_predict)
plt.plot([0,50],[0,50],'r')
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
```

## subplot畫多張圖
subplot(列數,行數,第n張圖)

```python
x = np.linspace(-10,10,200)
```

```python
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.sin(x))

plt.subplot(2,2,3)
plt.plot(x,np.sin(x))

plt.subplot(2,2,4)
plt.plot(x,np.sin(x))
```

## 所以可以彙整成這樣

```python
plt.figure(figsize=(8,10))#調整圖的尺寸、比例
for i,feature in enumerate(boston.feature_names):
    plt.subplot(5,3,i+1)#設定 n*m 的圖
    plt.scatter(X[:,i],Y,s = 1)
    plt.xlabel(feature)
    plt.ylabel('price')
    plt.tight_layout()#調整堆疊，避免 label卡在一起
```
